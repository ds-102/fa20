```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from math import cos, sin
import numpy as np
```

```python
def x0s(ps):
  return [p[0] for p in ps]

def x1s(ps):
  return [p[1] for p in ps]

circle = [(cos(t), sin(t)) for t in np.arange(0, 2*np.pi, .001)]
```

```python
points = [(0.2, -0.3)]


for ii in range(1000):
  (x0, x1) = points[-1]
  if ii % 2 == 0: # x0 update
    t = (1 - x0**2)**0.5
    x0new = x0
    x1new = rnd.uniform(low = -t, high = t)
  else: # x1 update
    t = (1 - x1**2)**0.5
    x0new = rnd.uniform(low = -t, high = t)
    x1new = x1
  points.append((x0new, x1new))

plt.scatter(x0s(circle), x1s(circle), s=1, c='g', alpha=0.3, zorder=-10)
plt.scatter(x0s(points), x1s(points), c='b', alpha=0.5)
plt.scatter([x0], [x1], c='r', zorder=10)
plt.axis([-1,1,-1,1])
plt.axis('equal')
```
